```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("2024_short.csv", low_memory=False)
df['Created Date'] = pd.to_datetime(df['Created Date'], format="%m/%d/%Y %I:%M:%S %p", errors='coerce')
```

```python
start_early = "2024-01-01"
end_early = "2024-02-29"
df_early = df[(df['Created Date'] >= start_early) & (df['Created Date'] <= end_early)]
most_common_early = df_early['Complaint Type'].value_counts().idxmax()

df_plot_early = df_early[df_early['Complaint Type'] == most_common_early]
counts_borough = df_plot_early.groupby('Borough').size()

counts_borough.plot(kind='bar', color='skyblue')
plt.title(f"Most Common Complaint Type in Early 2024: {most_common_early}")
plt.xlabel("Borough")
plt.ylabel("Number of Complaints")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
june_start = "2024-06-01"
july_end = "2024-07-31"
df_june_july = df[(df['Created Date'] >= june_start) & (df['Created Date'] <= july_end)]
df_plot_summer = df_june_july[df_june_july['Complaint Type'] == most_common_early]
counts_borough_summer = df_plot_summer.groupby('Borough').size()
counts_borough_summer.plot(kind='bar', color='lightcoral')
plt.title(f"{most_common_early} in June-July 2024")
plt.xlabel("Borough")
plt.ylabel("Number of Complaints")  
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
